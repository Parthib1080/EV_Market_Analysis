```python
# Create categories based on cars sold

df['sales_category'] = pd.cut(
    df['annual_sales_units'],
    bins=[0, 50000, 100000, 250000, 400000, float('inf')],
    labels=[
        'Unpopular',
        'Sold',
        'Popular',
        'Best Seller',
        'Top Selling'
    ]
)
```

```python
df[['annual_sales_units','sales_category']].head(15)
```

```python
df['performance_type'] = pd.cut(
    df['acceleration_0_60_mph'],
    bins=[2, 4, 6, 8, float('inf')],
    labels=['Sports', 'Fast', 'Moderate', 'Slow']
)
```
